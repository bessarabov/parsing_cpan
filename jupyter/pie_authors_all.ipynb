```python
import matplotlib.pyplot as plt

%matplotlib inline

data = [5410, 1543, 51, 50]
labels = ['spaces & tabs', 'only spaces', 'only tabs', '']
colors = ['#481f6a', 'green', 'blue', 'silver']

fig, ax = plt.subplots()
fig.subplots_adjust(0.3, 0, 1.1, 1.1)
wedges, labels, autopct = ax.pie(data, 
                                 radius=3800,
                                 labels=labels,
                                 labeldistance=1.1,
                                 autopct='%1.1f%%', 
                                 colors=colors,
                                 startangle=90, 
                                 counterclock=False)
plt.setp(labels, fontsize=18)
ax.axis('equal')

plt.show()
```
