```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
df = pd.read_csv('releases_date.csv',parse_dates=['date'])
df.loc[:,'total'] = df.sum(axis=1)

df['00_percent'] = df['00'] * 100 / df['total']
df['01_percent'] = df['01'] * 100 / df['total']
df['10_percent'] = df['10'] * 100 / df['total']
df['11_percent'] = df['11'] * 100 / df['total']

df
```

```python
v00 = df.iloc[:, 1].values.reshape(-1, 1)
v01 = df.iloc[:, 2].values.reshape(-1, 1)
v10 = df.iloc[:, 3].values.reshape(-1, 1)
v11 = df.iloc[:, 4].values.reshape(-1, 1)

v00_percent = df.iloc[:, 6].values.reshape(-1, 1)
v01_percent = df.iloc[:, 7].values.reshape(-1, 1)
v10_percent = df.iloc[:, 8].values.reshape(-1, 1)
v11_percent = df.iloc[:, 9].values.reshape(-1, 1)
```

```python
plt.rcParams["figure.figsize"] = (20, 10)

fig, ax = plt.subplots()

ax.plot(df['date'], v10_percent, color='green', label='only spaces')
ax.plot(df['date'], v01_percent, color='blue', label='only tabs')
ax.plot(df['date'], v11_percent, color='#481f6a', label='spaces & tabs')
ax.plot(df['date'], v00_percent, color='silver', label='no spaces & no tabs')

ax.legend()

plt.show()
```
